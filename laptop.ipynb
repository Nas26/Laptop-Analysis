```python
import seaborn as sns
company_price_df = df.groupby('Company')['Price'].sum().reset_index()

plt.figure(figsize=(10, 8))
plt.pie(company_price_df['Price'], labels=company_price_df['Company'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Price Distribution by Laptop Company')
plt.show()
```

```python
price_by_company = df.groupby('Company')['Price'].sum().reset_index()

# Sort the DataFrame by price in descending order
price_by_company_sorted = price_by_company.sort_values(by='Price', ascending=False)

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(price_by_company_sorted['Company'], price_by_company_sorted['Price'], color='skyblue')
plt.xlabel('Company')
plt.ylabel('Total Price')
plt.title('Total Price Distribution by Laptop Company')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
